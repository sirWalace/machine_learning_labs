## 1. Вероятностное пространство, формула Байеса

### Задача 1
Определить (с обоснованием), зависимы или независимы следующие события:

(a) Несовместные события:


События A и B несовместны  $ \ \Leftrightarrow \  P(AB) = 0$ 


$P(A|B) = P(A)$ - независимость. $P(A|B) = \frac{P(A \cap B)}{P(B)} = \frac{P(AB)}{P(B)} = 0 \neq P(A)$

Значит несовместные события **зависимы**.

----

(b) События, образующие $\sigma$-алгебру $\Sigma$ в пространстве ($\Omega, \Sigma, \mathbb{P}$)

Тут не очень понятно что имеется ввиду.

(c) События, имеющие одинаковую вероятность
$P(A) = P(B)$

1. A - выпал орёл, B - выпала решка. Вероятности равны, но события зависимы.
2. A - выпал орёл на первой монете, B - выпала решка на второй монете. Вероятности равны, но события независимы.

Таким образом, события имеющие одинаковую вероятность могут быть как зависимыми, так и не зависимыми.

## 2. Случайный вектор и числовые характеристики

### Задача 1

Числитель и знаменатель не отрицательные $\Rightarrow \ \ f(x) > 0$. Значит если интеграл от этой функции равен 1, то найдётся случайные вектор для которого $f(x)$ будет являться плотностью. Проверим:

$$\int\limits_{\mathbb{R}^2} \frac{e^{-2|y|}}{\pi (1+x^2)} =  \frac{1}{\pi} \int\limits_{\mathbb{R}} e^{-2|y|} dy \cdot \int\limits_{\mathbb{R}} \frac{dx}{1+x^2} = \frac{2}{\pi} \int\limits_{0}^{\infty} e^{-2y} dy \cdot \int\limits_{\mathbb{R}} (\arctg (x))' dx =  $$

$$= \frac{2}{\pi} \cdot \frac{e^{-2y}}{2}\big|_{0}^{\infty}  \cdot (\frac{\pi}{2} + \frac{\pi}{2}) = \frac{2}{\pi} \cdot \frac{1}{2} \cdot \pi= 1$$

 $$\mathbb{D}(\xi) = \mathbb{E}(\xi^2) = 1; \ \ \ \ \mathbb{D}(\eta) = \mathbb{E}(\eta^2) - \mathbb{E}(\eta)^2  = \frac34 - \frac{1}{64} = \frac{47}{64} ; $$
