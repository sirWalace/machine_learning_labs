# Задача 1

Пусть $X_1, X_2, \ldots, X_n$ — выборка из экспоненциального распределения с параметром $\lambda$. Найти оценку максимального правдоподобия параметра $\lambda$, сравнить ее с байесовской оценкой (MAP и математическое ожидание апостреорного распределения), подобрав сопряженное распределение. Сравнить полученные байесовские оценки с оценкой MLE. Найти предсказательное распределение

#### Решение

**Оценка максимального правдоподобия:**

$$ L(X_1,...,X_n|\lambda) = f(X_1, \lambda) \cdot ... \cdot f(X_n, \lambda) = \lambda^n e^{-\lambda \sum\limits_{i=1}^n X_i} \ \Rightarrow $$

$$ \Rightarrow l(X_1,...,X_n|\lambda)  =  \ln L(X_1,...,X_n|\lambda) = n \ln \lambda - \lambda \sum\limits_{i=1}^n X_i \rightarrow \underset{\lambda}{\max} \ \Rightarrow $$

$$ \Rightarrow l'_{\lambda} = \frac{n}{\lambda} - \sum\limits_{i=1}^n X_i = 0 \ \Rightarrow \ \boxed{ \lambda_{MLE} = \frac{n}{\sum\limits_{i=1}^n X_i} = \frac{1}{\overline X}} $$

----

**Байесовская оценка:**

Для экспоненциального распределения сопряженным является Гамма.

$$ p(\lambda | X) \propto p(\lambda) \cdot p(X| \lambda) $$

$p( X | \lambda)$ - правдоподобие, мы уже вычислили.

$$ p(\lambda | X) \propto \lambda^n e^{-\lambda n \overline X} \cdot \lambda^{k-1} e^{-\lambda \beta} = \lambda^{n+k-1} e^{-\lambda (n\overline X + \beta)} $$

Получили что апостериорное распределение - Гамма распределение с параметрами $(n+k)$ и $(n\overline X + \beta)$

Теперь посчитаем MAP оценку. Процесс вывода аналогичен MLE(ищем максимум), так что можно сразу сказать ответ:


$$ \boxed{\lambda_{MAP} = \frac{n+k-1}{n\overline X + \beta} } $$


Если приравнять к нулю поправки связанные с априорным распределением, получим в точности оценку MLE.


----

**Предсказательное распределение:**

$$ p(X) = \int\limits_0^{\infty} p(X| \lambda) p (\lambda ) d \lambda = \int\limits_0^{\infty} \lambda^n e^{-\lambda n \overline X} \cdot \lambda^{k-1} e^{-\lambda \beta} d \lambda = \int\limits_0^{\infty} \lambda^{n+k-1} e^{-\lambda (n\overline X + \beta)}  d \lambda = \left[ y = \lambda (n\overline X + \beta) \right]  = $$

$$ = (n\overline X + \beta)^{-1} \cdot (n\overline X + \beta)^{1-n-k}  \int\limits_0^{\infty} y^{n+k-1} e^{-y}  dy = \frac{1}{(n\overline X + \beta)^{n+k}}  \Gamma(n+k) =\boxed{ \frac{(n+k-1)!}{(n\overline X + \beta)^{n+k}} }$$

-----

**Мат. ожидание апостериорного распределения:**

$$ \mathbb{E}( \Lambda ) = \left( \frac{(n+k-1)!}{(n\overline X + \beta)^{n+k}} \right)^{-1} \cdot \int\limits_0^{\infty} \lambda \lambda^{n+k-1} e^{-\lambda (n\overline X + \beta)}  d\lambda = \left( \frac{(n+k-1)!}{(n\overline X + \beta)^{n+k}} \right)^{-1} \cdot \int\limits_0^{\infty}  \lambda^{n+k} e^{-\lambda (n\overline X + \beta)} d \lambda = \left( \frac{(n+k-1)!}{(n\overline X + \beta)^{n+k}} \right)^{-1}  \frac{(n+k)!}{(n\overline X + \beta)^{n+k+1}} = $$


$$ = \boxed{ \frac{n+k}{n\overline X + \beta} } $$

# Задача 2

**Мультиномиальное распределение**

Пусть проводится серия из $n$ испытаний и в результате каждого испытания происходит ровно одно событие из набора $A_1, A_2, \dots, A_m$, причем вероятности этих событий равны соответственно $\mathsf{p}_1, \mathsf{p}_2, \dots, \mathsf{p}_m$, причем
$$\sum_{i=1}^{m}\mathsf{p}_i = 1.$$

Тогда совместное распределение величин $X_1, X_2, \dots, X_m$, где $X_k$ — число наступлений события $A_k$ в серии из $n$ испытаний, задается вероятностями

$$
\mathsf{P}\left(X_1 = n_1, \dots, X_m = n_m, \right) = \frac{n!}{n_1!\dots n_m!}\mathsf{p}_1^{n_1}\dots \mathsf{p}_m^{n_m},
$$

где $n_1, n_2, \dots, n_m$ — произвольный набор целых неотрицательных чисел, таких что

$$\sum_{i=1}^m n_i = n.$$

Произведите байесовский вывод для мультиномиального распределения: найдите апостериорное распределение, используя в качестве сопряженного распределения к правдоподобию [распределение Дирихле](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%94%D0%B8%D1%80%D0%B8%D1%85%D0%BB%D0%B5), найдите предсказательное распределение. Объясните результат.

#### Решение

**Функция правдоподобия**

$$L(X_1,..., X_m| \vec p) =p_1^{X_1}...p_m^{X_m} $$

Вероятность распределения Дирихле имеет вид:

$$ p(X_1,..., X_m| \vec p) = \frac{1}{\Beta(X)} p_1^{\alpha_1-1}...p_m^{\alpha_m-1}  $$

Тогда можем записать(с учётом распределения Дирихле) апостериорное распределение:

$$ \boxed{ P(\vec p | X_1,...,X_m) \propto  \prod\limits_{i=1}^m p_i^{X_i + \alpha_i -1} \sim Dir(\vec a + \vec X) } $$

Найдём максимум с учетом $\sum\limits_{j=1}^m p_i = 1$:

$$ l = \ln ( \prod\limits_{i=1}^m p_i^{X_i + \alpha_i -1}) +\lambda (\sum\limits_{i=1}^m p_i - 1) = \sum\limits_{i=1}^m (X_i + \alpha_i -1)\ln (  p_i) +\lambda (\sum\limits_{i=1}^m p_i - 1) $$

$$l'_{p_i} = \frac{X_i + \alpha_i -1}{p_i} + \lambda = 0 \ \ \Rightarrow \ \ p_i = - \frac{X_i + \alpha_i -1}{\lambda}$$

$$l'_{\lambda} = \sum\limits_{i=1}^m p_i - 1 = 0 \ \ \Rightarrow \ \ \sum\limits_{i=1}^m p_i = 1 \ \ \Rightarrow \ \ - \frac{\sum X_i + \sum \alpha_i -\sum 1}{\lambda} = - \frac{n+ \sum \alpha_i - m}{\lambda} = 1 \ \ \Rightarrow \ \ \lambda =-n+ \sum \alpha_i - m $$

Итого:

$$ \boxed{ p_i^{MAP} = \frac{X_i + \alpha_i - 1}{n + \sum \alpha_i - m}  } $$

Заметим что при $\alpha_i = 1$(отсутствие изначальных предположений), MAP оценка становится равна MLE оценке для мультиномиального распределения
