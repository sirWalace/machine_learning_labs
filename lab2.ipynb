(a)

$$ f_{\xi}(x) = \theta^2 x e^{-\theta x} \ \Rightarrow \ \mathbb{E}(\xi) = \int\limits_{0}^{\infty} x f_{\xi}(x) dx = \theta^2 \int\limits_{0}^{\infty}  x^2 e^{-\theta x} dx =\left[ \theta x =  y \right] = $$

$$ = \frac{1}{\theta} \int\limits_{0}^{\infty} y^2 e^{-y} dy = \frac{1}{\theta} \Gamma(3) =\boxed{ \frac{2}{\theta} } $$

$$ \mathbb{E}(\xi^2) = \theta^2 \int\limits_{0}^{\infty}  x^3 e^{-\theta x} dx = \frac{1}{\theta^2} \int\limits_{0}^{\infty} y^3 e^{-y} dy = \frac{\Gamma(4)}{\theta^2} = \boxed{ \frac{6}{\theta^2} } $$

$$\mathbb{D}(\xi) = \mathbb{E}(\xi^2) - \mathbb{E}(\xi)^2 = \frac{6}{\theta^2} - \frac{4}{\theta^2} = \boxed{\frac{2}{\theta^2} } $$


\begin{array}{c|c|c|c}
\theta  & \mathbb{E}(\xi) & \mathbb{E}(\xi^2) & \mathbb{D}(\xi) \\ \hline
0.5 & 4 & 24 & 8 \\
2 & 1 & 1.5 & 0.5 \\
8 & 0.25 & \frac{3}{32} & \frac{1}{32}
\end{array}

### Задача 2 

(a)

**Мода:**
$$ f_{\xi}'(x) = -\lambda^2 e^{-\lambda(x-a)} = -4 e^{-2(x-2)} < 0 $$

Т.е. плотность убывает, а значит мода(самое частое значение в выборке) = a = 2


**Мат. ожидание:**
$$ \mathbb{E}(\xi) = \int\limits_{\mathbb{R}} x f_{\xi}(x) dx = 2 \int\limits_{2}^{\infty} x e^{-2(x-2)} dx =\boxed{ 2.5} $$


**Медиана:**
$$ F_{\xi}(x) = P(\xi < x) = \int\limits_{2}^x f_{\xi}(x) dx = 0.5 = - e^{-2(x-2)}\big|^x_2 = - \left( e^{-2(x-2)} - 1 \right) \Rightarrow  $$

$$ \Rightarrow \ e^{-2(x-2)} = 0.5 \ \Rightarrow \ -2(x-2) = \ln 0.5 \ \Rightarrow \ \boxed{ x =  2 - \frac12 \ln \frac12 \approx 2.35 }  $$



(b)

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Создание пользовательского непрерывного распределения
class customDist(stats.rv_continuous):
    def _pdf(self, x):
        # Ваша функция плотности
        return 0 if x < 2 else 2*np.exp(-2*(x -2)) 

# Создание экземпляра распределения
my_dist = customDist()

# Генерация выборки
sample_large = my_dist.rvs(size=10000)
sample_small = my_dist.rvs(size=20)
```

```python
print(f'Выборочное среднее большой выборки: {round(sample_large.mean(), 2)}')
print(f'Выборочное среднее маленькой выборки: {round(sample_small.mean(), 2)}\n')

print('---------------')

print(f'Выборочная медиана большой выборки: {round(np.median(sample_large), 2)}')
print(f'Выборочная медиана маленькой выборки: {round(np.median(sample_small), 2)}\n')

print('---------------')
print(f'Выборочная мода большой выборки: {round(sp.stats.mode(sample_large)[0], 2)}')
print(f'Выборочная мода маленькой выборки: {round(sp.stats.mode(sample_small)[0], 2)}')
```

### 2. Моделирование совместного распределения двух СВ

Вычислим маргинальные распределения:

$$ p(\xi = -1) = p(\xi = 1) = \frac25 \sum\limits_{i=1}^{\infty} \left( \frac12 \right)^i = \frac25 \frac{0.5}{1-0.5} = \boxed{\frac25} $$

$$ \boxed{p(\xi = 0) = \frac15} $$


$$ p(\eta = 1) = (\frac25 + \frac15 + \frac25) \frac12 = \boxed{\frac12} \ ; \ \ \ p(\eta = 2 ) = (\frac25 + \frac15 + \frac25) \left( \frac12 \right)^2 =  \left( \frac12 \right)^2 $$

Общее правило: $$ \boxed{ p(\eta = n) = \left( \frac12 \right)^n } $$


$$ \mathbb{E}(\xi) = -1 \cdot \frac25 + 1 \cdot \frac25 + 0 \cdot \frac15 = 0  \ ; \ \ \ \mathbb{E}(\xi^2) =  (-1)^2 \cdot \frac25 + 1^2 \cdot \frac25 + 0 \cdot \frac15 = \frac45 \ ; \ \ \ \mathbb{D}(\xi) = \frac45 $$


$$ \mathbb{E}(\eta) = \sum\limits_{n=1}^{\infty} n \left( \frac12 \right)^n = \sum\limits_{n=1}^{\infty} \frac{0.5^n}{0.5} = \sum\limits_{n=1}^{\infty} \left( \frac12 \right)^{n-1}=2\ ; \ \ \ \mathbb{E}(\eta^2) = \sum\limits_{n=1}^{\infty} n^2 \left( \frac12 \right)^n = 6 \ ; \ \ \ \mathbb{D}(\eta) = 2$$


Ковариация:

$$ cov(\xi, \eta) = \sum\limits_{i,j}\left(\xi_i - \mathbb{E}(\xi)\right)\left(\eta_j - \mathbb{E}(\eta)\right) \cdot p(\xi = \xi_i, \ \eta = \eta_j) = \sum\limits_{i,j} \xi_i\left(\eta_j - 2 \right) \cdot p(\xi = \xi_i, \ \eta = \eta_j) = \sum\limits_{j} 0 \cdot \left(\eta_j - 2 \right) \cdot p(\xi = 0, \ \eta = \eta_j) +  $$

$$ + \sum\limits_{j} 1 \cdot \left(\eta_j - 2 \right) \cdot p(\xi = 1, \ \eta = \eta_j) - \sum\limits_{j} 1 \cdot \left(\eta_j - 2 \right) \cdot p(\xi = -1, \ \eta = \eta_j) = 0 $$

Итого, корреляционная матрица равна:
$$ \Sigma = \sqrt{\frac58} \begin{pmatrix} \frac45 & 0 \\ 0 & 2 \end{pmatrix} $$

```python
# Численно оценим ковариацонную матрицу
# Зададим распределение Кси
vals_xi = np.array([-1, 0, 1])
probs_xi = np.array([2/5, 1/5, 2/5])
discrete_dist_xi = np.random.choice(vals_xi, p=probs_xi, size=100000)


# Зададим распределение Эта
vals_eta = np.linspace(1, 1000, num=1000)
probs_eta = np.array([2 ** (-i) for i in range(1, 1001)])
discrete_dist_eta = np.random.choice(vals_eta, p=probs_eta, size=100000)

# Численные оценки
e_xi = discrete_dist_xi.mean()
e_eta = discrete_dist_eta.mean()

d_xi = discrete_dist_xi.var()
d_eta =  discrete_dist_eta.var()


cov_matrix = np.cov(discrete_dist_xi, discrete_dist_eta)

print(cov_matrix)
```
